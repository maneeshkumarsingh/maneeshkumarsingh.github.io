```python
df=df.drop('lastSeen',axis=1)
df=df.drop('postalCode',axis=1)
df=df.drop('dateCrawled',axis=1)
df=df.drop('dateCreated',axis=1)
df=df.drop('seller', axis=1)
df=df.drop('name', axis=1)
df.head()
```

```python
pz=pd.get_dummies(df.fuelType,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('fuelType',axis=1,inplace=True)

pz=pd.get_dummies(df.model,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('model',axis=1,inplace=True)

pz=pd.get_dummies(df.gearbox,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('gearbox',axis=1,inplace=True)


pz=pd.get_dummies(df.offerType,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('offerType',axis=1,inplace=True)

pz=pd.get_dummies(df.abtest,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('abtest',axis=1,inplace=True)

pz=pd.get_dummies(df.vehicleType,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('vehicleType',axis=1,inplace=True)

pz=pd.get_dummies(df.notRepairedDamage,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('notRepairedDamage',axis=1,inplace=True)

pz=pd.get_dummies(df.brand,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('brand',axis=1,inplace=True)
```

```python
df.head()
```

```python
labels = np.array(df['price'])
```

```python
df= df.drop('price', axis = 1)
```

```python
feature_list = list(df.columns)
```

```python
features = np.array(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 1000, random_state = 40)
```

```python
rf.fit(train_features, train_labels)
```

```python
predictions = rf.predict(test_features)
```

```python
errors = abs(predictions - test_labels) 
```

```python
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
mape = 100 * (errors / test_labels)
```

```python
accuracy = 100 - np.mean(mape)
```

```python
print('Accuracy:', round(accuracy, 2), '%.')
```
