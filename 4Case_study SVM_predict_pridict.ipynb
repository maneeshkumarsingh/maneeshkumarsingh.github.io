```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline
import math
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import std
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVR

#import seaborn as sns; sns.set(fornt_scale=1.2)
```

```python
df = pd.read_csv('cars_sample-modified.csv')
```

```python
df=df.drop('lastSeen',axis=1)
df=df.drop('postalCode',axis=1)
df=df.drop('dateCreated',axis=1)
df=df.drop('dateCrawled',axis=1)
df=df.drop('model',axis=1)
df=df.drop('name',axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df['vehicleType'] = df['vehicleType'].fillna((df['vehicleType'].mode()[0]))
df['gearbox'] = df['gearbox'].fillna((df['gearbox'].mode()[0]))
#df['model'] = df['model'].fillna((df['model'].mode()[0]))
df['fuelType'] = df['fuelType'].fillna((df['fuelType'].mode()[0]))
df['notRepairedDamage'] = df['notRepairedDamage'].fillna((df['notRepairedDamage'].mode()[0]))
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#identify outlier for price
dm, ds = mean(df.price), std(df.price)
cf = ds * 3
lower, upper = dm- cf, dm + cf
df[(df.price>upper)|(df.price<lower)]
```

```python
#identify outlier for powerPS
dm, ds = mean(df.powerPS), std(df.powerPS)
cf = ds * 3
lower, upper = dm- cf, dm + cf
df[(df.powerPS>upper)|(df.powerPS<lower)]
df.head()
```

```python
#identify outlier for kilometer
dm, ds = mean(df.kilometer), std(df.kilometer)
cf = ds * 3
lower, upper = dm- cf, dm + cf
df[(df.kilometer>upper)|(df.kilometer<lower)]
df.head()
```

```python
#for change dummy variyebel
pz=pd.get_dummies(df.fuelType,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('fuelType',axis=1,inplace=True)
pz=pd.get_dummies(df.seller,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('seller',axis=1,inplace=True)
pz=pd.get_dummies(df.offerType,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('offerType',axis=1,inplace=True)
pz=pd.get_dummies(df.abtest,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('abtest',axis=1,inplace=True)
pz=pd.get_dummies(df.vehicleType,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('vehicleType',axis=1,inplace=True)
pz=pd.get_dummies(df.notRepairedDamage,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('notRepairedDamage',axis=1,inplace=True)
pz=pd.get_dummies(df.brand,drop_first=True)
df=pd.concat([df,pz],axis=1)
df.drop('brand',axis=1,inplace=True)
```

```python
x=df.drop('gearbox',axis=1)
y=df['gearbox']
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=109) # 70% training and 30% test
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)
 
#Predict the response for test dataset
y_pred = clf.predict(X_test)
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
